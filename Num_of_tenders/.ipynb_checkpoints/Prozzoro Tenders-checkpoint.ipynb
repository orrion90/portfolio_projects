```python
import pandas as pd
```

```python
source_df = pd.read_excel('tenders_lots2.xlsx')
```

```python
df = source_df[:]
```

```python
df.head(1)
```

```python
df = df.rename(columns = {'Ідентифікатор лота': 'lot_id',
                     'Лот': 'lot',
                     'Очікувана вартість': 'expected_cost',
                     'Процедура закупівлі': 'procurement_procedure',
                     'Статуси процедур': 'status_procedures',
                     'Статус лота': 'lot_status',
                     '...': 'web_site',
                     'Організатор' : 'organizer',
                     'Переможець' : 'winner',
                     'Ціна за одиницю': 'price_per_unit',
                     'Одиниця виміру': 'unit_of_measure',
                     'Класифікація CPV': 'classification_cpv',
                     'Кіль-ть питань': 'num_questions',
                     'Кіль-ть унікальних учасників лота': 'num_of_unique_participants',
                     'Кількість дискваліфікацій лота': 'num_of_disq',
                     'Кількість скарг лота': 'num_of_complaints',
                     'Дата публікації процедури': 'date_of_publication_of_the_procedure'})
```

# Для каждего организатора тендера определяем его самого частого победителя

```python
df.lot_status.unique()
```

Судя по всему в данные попали неуспешные и отмененные тендера, нужно от них избавится

```python
df_drop = df.query('lot_status in ["Неуспішний лот закупівлі", "Скасовано лот закупівлі"]').index
df.drop(df_drop, axis=0, inplace=True)
```

```python
df.shape[0]
```

```python
df.isna().sum()
```

```python
res_df = df \
    .groupby(['organizer', 'winner'], as_index=False) \
    .aggregate({'lot_id': 'count'}) \
    .sort_values('lot_id', ascending=False)
```

```python
res_df
```

```python
res_df.drop_duplicates(subset = ['organizer'], inplace=True)
```

```python
res_df.rename(columns={'organizer': 'Організатор', \
                       'winner': 'Переможець', \
                       'lot_id': 'Кількість виграних лотів'}, inplace=True)
```

```python
res_df
```

# Вывести общий бюджет лотов, кол-во лотов, кол-во тендеров, по обобщённый категории CPV до 2 знака

```python
df.head()
```

Id тендера содержится в названии лота

```python
df['tenders'] = df.lot_id.apply(lambda x: x[:22])
```

```python
df['tenders']
```

Обобщаем категорию CPV до 2 знаков

```python
df.classification_cpv = df.classification_cpv.apply(lambda x: x[:2])
```

```python
num_of_tenders = df \
    .groupby(['classification_cpv'], as_index=False) \
    .aggregate({'expected_cost': 'sum',
                'lot_id': 'count', \
                'tenders': 'nunique'}) \
    
```

```python
 num_of_tenders.expected_cost = num_of_tenders.expected_cost / 1000000
```

```python
num_of_tenders.rename(columns = {
                'classification_cpv': 'Узагальнені категорії CPV',
                'expected_cost': 'Загальний очікуваний бюджет лотів (млн.грн.)',
                'lot_id': 'Кількість лотів',
                'tenders': 'Кількість тендерів'}, inplace=True)
```

```python
num_of_tenders.head()
```

# Вывести сводную помесячную таблицу, в которой указать кол-во лотов, кол-во торгов, кол-во дисквалификаций и кол-во жалоб

```python
df.head()
```

```python
df.date_of_publication_of_the_procedure.min()
```

```python
df.date_of_publication_of_the_procedure.max()
```

```python
df['month_of_publication'] = df['date_of_publication_of_the_procedure'].dt.month
```

```python
df.head(1)
```

```python
num_of_complaints = df \
    .groupby('month_of_publication', as_index=False) \
    .aggregate({'lot_id': 'count',
                'tenders': 'nunique', \
                'num_of_disq': 'sum', \
                'num_of_complaints': 'sum'}) \
    .sort_values('month_of_publication') \
    .rename(columns = {
                'month_of_publication': 'Місяць проведення торгів',
                'lot_id': 'Кількість лотів',
                'tenders': 'Кількість тендерів',
                'num_of_disq': 'Кількість дискваліфікацій',
                'num_of_complaints': 'Кількість скарг'})
num_of_complaints
```
