```python
import pandas as pd
```

```python
source_df = pd.read_excel('tenders_lots2.xlsx')
```

```python
df=source_df
```

```python
df = df.rename(columns = {'Ідентифікатор лота': 'lot_id',
                     'Лот': 'lot',
                     'Очікувана вартість': 'expected_cost',
                     'Процедура закупівлі': 'procurement_procedure',
                     'Статуси процедур': 'status_procedures',
                     'Статус лота': 'lot_status',
                     '...': 'web_site',
                     'Організатор' : 'organizer',
                     'Переможець' : 'winner',
                     'Ціна за одиницю': 'price_per_unit',
                     'Одиниця виміру': 'unit_of_measure',
                     'Класифікація CPV': 'classification_cpv',
                     'Кіль-ть питань': 'num_questions',
                     'Кіль-ть унікальних учасників лота': 'num_of_unique_participants',
                     'Кількість дискваліфікацій лота': 'num_of_disq',
                     'Кількість скарг лота': 'num_of_complaints',
                     'Дата публікації процедури': 'date_of_publ_of_the_proc'})
```

```python
df.head(3)
```

```python
df['date_of_publ_of_the_proc'] = df['date_of_publ_of_the_proc'].dt.month
```

```python
df.head()
```

```python
def changer (value):
    value_new = value[:22]
    return value_new
```

```python
df['num_of_tender'] = df.lot_id.apply(changer)
```

```python
df.num_of_tender = df.num_of_tender.drop_duplicates()
```

```python
fin_df = df \
    .groupby('date_of_publ_of_the_proc', as_index=False) \
    .aggregate({'lot_id': 'count',
                'num_of_tender': 'count', \
                'num_of_disq': 'sum', \
                'num_of_complaints': 'sum'}) \
    .sort_values('date_of_publ_of_the_proc') \
    .rename(columns = {
                'date_of_publ_of_the_proc': 'Місяць проведення торгів',
                'lot_id': 'Кількість лотів',
                'num_of_tender': 'Кількість тендерів',
                'num_of_disq': 'Кількість дискваліфікацій',
                'num_of_complaints': 'Кількість скарг'})
```
