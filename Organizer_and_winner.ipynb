```python
import pandas as pd
```

```python
source_df = pd.read_excel('tenders_lots2.xlsx')
```

```python
df = source_df
```

```python
df = df.rename(columns = {'Ідентифікатор лота': 'lot_id',
                     'Лот': 'lot',
                     'Очікувана вартість': 'expected_cost',
                     'Процедура закупівлі': 'procurement_procedure',
                     'Статуси процедур': 'status_procedures',
                     'Статус лота': 'lot_status',
                     '...': 'web_site',
                     'Організатор' : 'organizer',
                     'Переможець' : 'winner',
                     'Ціна за одиницю': 'price_per_unit',
                     'Одиниця виміру': 'unit_of_measure',
                     'Класифікація CPV': 'classification_cpv',
                     'Кіль-ть питань': 'num_questions',
                     'Кіль-ть унікальних учасників лота': 'num_of_unique_participants',
                     'Кількість дискваліфікацій лота': 'num_of_disq',
                     'Кількість скарг лота': 'num_of_complaints',
                     'Дата публікації процедури': 'date_of_publication_of_the_procedure'})
```

```python
df['organizer'].unique().shape[0]
```

```python
filt = (df['winner'].apply(len) < 2)
```

```python
df[filt].groupby('lot_status').count()
```

```python
filt = (df['winner'].apply(len) < 2)
df.drop(df[filt].index, inplace=True)
```

```python
df.shape[0]
```

```python
df.isna().sum()
```

```python
res_df = df \
    .groupby(['organizer', 'winner'], as_index=False) \
    .aggregate({'lot_id': 'count'}) \
    .sort_values('lot_id', ascending=False)
```

```python
res_df
```

```python
res_df.drop_duplicates(subset = ['organizer'], inplace=True)
```

```python
res_df.rename(columns={'organizer': 'Організатор', \
                       'winner': 'Переможець', \
                       'lot_id': 'Кількість виграних лотів'}, inplace=True)
```

```python
res_df
```

```python
res_df.to_excel('task1_corr3.xlsx', index=False)
```
